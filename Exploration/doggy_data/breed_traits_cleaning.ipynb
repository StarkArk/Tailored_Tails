```python
import pandas as pd
```

```python
#read in csv
df = pd.read_csv('akc-data-latest.csv')
df.head(50)
```

```python
#filter top 50 breeds based on 2022 AKC data
breeds = [
    'French Bulldog', 'Labrador Retriever', 'Golden Retriever', 
    'German Shepherd Dog', 'Poodle (Standard)', 'Bulldog', 'Rottweiler', 
    'Beagle', 'Dachshund', 'German Shorthaired Pointer', 
    'Pembroke Welsh Corgi', 'Australian Shepherd', 'Yorkshire Terrier', 
    'Cavalier King Charles Spaniel', 'Doberman Pinscher', 'Boxer', 
    'Miniature Schnauzer', 'Cane Corso', 'Great Dane', 'Shih Tzu', 
    'Siberian Husky', 'Bernese Mountain Dog', 'Pomeranian', 
    'Boston Terrier', 'Havanese', 'English Springer Spaniel', 
    'Shetland Sheepdog', 'Brittany', 'Cocker Spaniel', 'Border Collie', 
    'Miniature American Shepherd', 'Belgian Malinois', 'Vizsla', 
    'Chihuahua', 'Pug', 'Basset Hound', 'Mastiff', 'Maltese', 'Collie', 
    'English Cocker Spaniel', 'Rhodesian Ridgeback', 'Newfoundland', 
    'Shiba Inu', 'Weimaraner', 'West Highland White Terrier', 
    'Portuguese Water Dog', 'Bichon Frise', 'Australian Cattle Dog', 
    'Dalmatian', 'Bloodhound'
]
df_50 = df[df['Unnamed: 0'].isin(breeds)]
df_50.head()
```

```python
#drop columns
columns_to_drop = ['popularity', 'grooming_frequency_category', 'shedding_category', 'group', 'energy_level_category', 'trainability_category', 'demeanor_category']
df_50_filtered = df_50.drop(columns=columns_to_drop)


df_50_filtered.head()
```

```python
#average max & min to get general info for various categories, and drop original columns to simplify data
df_50_filtered['average_height'] = df_50_filtered[['min_height', 'max_height']].mean(axis=1)
df_50_filtered['average_expectancy'] = df_50_filtered[['min_expectancy', 'max_expectancy']].mean(axis=1)
df_50_filtered['average_weight'] = df_50_filtered[['min_weight', 'max_weight']].mean(axis=1)
columns_to_drop = ['min_height', 'max_height', 'min_weight', 'max_weight', 'min_expectancy', 'max_expectancy']
df_50_filtered.drop(columns=columns_to_drop, inplace=True)
df_50_filtered.rename(columns={'Unnamed: 0':'Breed'}, inplace=True)
# df_50_filtered.set_index('Breed', inplace=True)
df_50_filtered
```

```python
# set size categories based on average weight from AKC standards
def map_size(average_weight):
    if average_weight > 75:
        return 1.0
    elif 55 <= average_weight <= 75:
        return 0.8
    elif 35 <= average_weight < 55:
        return 0.6
    elif 9 <= average_weight < 35:
        return 0.4
    else:
        return 0.2
df_50_filtered['size'] = df_50_filtered['average_weight'].apply(map_size)
df_50_filtered
```

```python
df_50_filtered.to_csv('top50breeds.csv', index=False)
```
